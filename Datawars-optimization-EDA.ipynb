```python
## Creating a list of numerical data features 
lst_numerical = []
for i in df.columns:
    if(df[i].dtype==np.float64 or df[i].dtype==np.int64):
        lst_numerical.append(i)
print(lst_numerical)
```

```python
df[lst_numerical].agg(['min','max','mean'])
```

```python
df[lst_numerical].describe()
```

```python
## Creating a list for object datatype features 
lst_object = []
for i in df.columns:
    if(df[i].dtype == 'object'):
        lst_object.append(i)
print(lst_object)
```

```python
df_2['Date'] = pd.to_datetime(df_2['Month'] + ' ' + df_2['Year'].astype(str))

# get quarter from date
df_2['Quarter'] = df_2['Date'].dt.to_period('Q')
```
