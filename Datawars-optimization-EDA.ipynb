```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df = pd.read_csv('datawars.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
## Creating a list of numerical data features 
lst_numerical = []
for i in df.columns:
    if(df[i].dtype==np.float64 or df[i].dtype==np.int64):
        lst_numerical.append(i)
print(lst_numerical)
```

```python
df[lst_numerical].agg(['min','max','mean'])
```

```python
df[lst_numerical].describe()
```

```python
## Creating a list for object datatype features 
lst_object = []
for i in df.columns:
    if(df[i].dtype == 'object'):
        lst_object.append(i)
print(lst_object)
```
